```python
import json
import os
from itertools import chain
from typing import Generator, Any
```

```python
path = r"C:\Users\Guilh\Videos\teste\root\workspace\data-eng-test\jsons"
```

```python
def json_files(*args, **kwargs) -> None:
    def reader_files() -> Generator[dict[str, Any], Any, None]:
        for arg in args:
            try:
                with open(arg) as f:
                    data = json.load(f)
                    chain_file = chain(data['assortment'])
                    product = next(chain_file)
                    dict_sku = {
                        "idRetailerSKU":product.get('idRetailerSKU',None),
                        "seller":product.get('seller',None),
                        "data":product.get('dateTimeReference',None)}
                    yield dict_sku

            except Exception as e:
                print(e)
    
    def create_dicts(*args, **kwargs) -> None:
        print([{**arg} for arg in args])
        
          
    dicts = reader_files()
    create_dicts(next(dicts))

jsonfiles = os.listdir(path)
for jsons in jsonfiles:
    json_file = os.path.join(path,jsons)
    json_files(json_file)
```
